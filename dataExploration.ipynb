```python
import pandas as pd

# Laden des Datensatzes
file_path = 'Ad_table (extra).csv'
data = pd.read_csv(file_path)

# Überprüfen der Datentypen
print(data.dtypes)

# Statistische Zusammenfassung der numerischen Merkmale
numerical_summary = data.describe()
print("Statistische Zusammenfassung der numerischen Merkmale:")
print(numerical_summary)

# Häufigkeiten der kategorialen Merkmale
categorical_summary = data.select_dtypes(include=['object']).apply(pd.Series.value_counts)
print("Häufigkeiten der kategorialen Merkmale:")
print(categorical_summary)
```

```python
import matplotlib.pyplot as plt

# Histogramme für numerische Merkmale
data.hist(bins=50, figsize=(20,15))
plt.show()

# Balkendiagramme für kategoriale Merkmale
for column in data.select_dtypes(include=['object']):
    data[column].value_counts().plot(kind='bar')
    plt.title(column)
    plt.show()
```

```python
# Umwandlung von numerischen Merkmalen, die als Objekte gespeichert sind
# Nur wenn der Wert ein String ist, führen wir die Umwandlung durch
data['Runned_Miles'] = pd.to_numeric(data['Runned_Miles'].apply(lambda x: x.replace(',', '') if isinstance(x, str) else x), errors='coerce')
data['Engin_size'] = pd.to_numeric(data['Engin_size'].apply(lambda x: x if pd.isna(x) else x), errors='coerce')
data['Annual_Tax'] = pd.to_numeric(data['Annual_Tax'].apply(lambda x: x.replace('£', '') if isinstance(x, str) else x), errors='coerce')
data['Average_mpg'] = pd.to_numeric(data['Average_mpg'].apply(lambda x: x.replace(' mpg', '') if isinstance(x, str) else x), errors='coerce')
data['Top_speed'] = pd.to_numeric(data['Top_speed'].apply(lambda x: x.replace(' mph', '') if isinstance(x, str) else x), errors='coerce')

# Boxplot für jedes numerische Merkmal in einer Matrix erstellen
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Anzahl der Spalten für die Matrix
num_cols = 4
num_rows = len(numerical_columns) // num_cols + (len(numerical_columns) % num_cols > 0)

fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()

for i, column in enumerate(numerical_columns):
    data.boxplot(column=[column], ax=axes[i])
    axes[i].ticklabel_format(style='plain', axis='y')  # Deaktivieren der wissenschaftlichen Notation
    axes[i].set_title(f'Boxplot für {column}')

# Entfernen leerer Subplots
for i in range(len(numerical_columns), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()

# Speichern der Boxplot-Matrix in eine Datei
output_file = 'boxplot_matrix.png'
plt.savefig(output_file)

plt.show()

print(f'Die Boxplot-Matrix wurde als Datei unter {output_file} gespeichert.')

# Die engin_size wurde hier im Boxplot leer angezeigt, was auf eine falsche Umwandlung der 
```

```python
import pandas as pd

# Laden des Datensatzes
file_path = 'Ad_table (extra).csv'
data = pd.read_csv(file_path)

# Entfernen von Duplikaten
data_cleaned = data.drop_duplicates()

# Imputation fehlender Werte (Beispiel: Forward Fill)
data_cleaned = data_cleaned.fillna(method='ffill')

# Konvertieren von numerischen Merkmalen, die als 'object' gespeichert sind
data_cleaned['Runned_Miles'] = pd.to_numeric(data_cleaned['Runned_Miles'], errors='coerce')
data_cleaned['Engin_size'] = pd.to_numeric(data_cleaned['Engin_size'], errors='coerce')
data_cleaned['Annual_Tax'] = pd.to_numeric(data_cleaned['Annual_Tax'], errors='coerce')
data_cleaned['Average_mpg'] = pd.to_numeric(data_cleaned['Average_mpg'], errors='coerce')
data_cleaned['Top_speed'] = pd.to_numeric(data_cleaned['Top_speed'], errors='coerce')

# Überprüfen von Wertebereichen
data_cleaned = data_cleaned[(data_cleaned['Runned_Miles'] >= 0) & (data_cleaned['Runned_Miles'] <= 1_000_000)]

# Konsistente Formatierung
data_cleaned['Color'] = data_cleaned['Color'].str.lower().str.strip()

# Validierung von Kategorien
valid_fuel_types = ['petrol', 'diesel', 'hybrid', 'electric']
data_cleaned = data_cleaned[data_cleaned['Fuel_type'].isin(valid_fuel_types)]

# Anzeigen der bereinigten Daten
print(data_cleaned.head())
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2_contingency

# Laden des Datensatzes
file_path = 'Ad_table (extra).csv'
data = pd.read_csv(file_path)

# Überprüfung der Daten vor der Umwandlung
print("Vor der Umwandlung:")
print(data['Engin_size'].unique())

# Konvertieren von numerischen Merkmalen, die als 'object' gespeichert sind
data['Runned_Miles'] = pd.to_numeric(data['Runned_Miles'].apply(lambda x: x.replace(',', '') if isinstance(x, str) else x), errors='coerce')
data['Engin_size'] = pd.to_numeric(data['Engin_size'].apply(lambda x: x.replace('L', '').strip() if isinstance(x, str) else x), errors='coerce')
data['Annual_Tax'] = pd.to_numeric(data['Annual_Tax'].apply(lambda x: x.replace('£', '') if isinstance(x, str) else x), errors='coerce')
data['Average_mpg'] = pd.to_numeric(data['Average_mpg'].apply(lambda x: x.replace(' mpg', '') if isinstance(x, str) else x), errors='coerce')
data['Top_speed'] = pd.to_numeric(data['Top_speed'].apply(lambda x: x.replace(' mph', '') if isinstance(x, str) else x), errors='coerce')

# Überprüfung der Daten nach der Umwandlung
print("Nach der Umwandlung:")
print(data['Engin_size'].unique())

# Auswahl der numerischen Merkmale
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

# Berechnung der Korrelationsmatrix
correlation_matrix = data[numerical_columns].corr()

# Erstellung der Heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap der Korrelationsmatrix')
plt.savefig('correlation_heatmap.png')  # Speichern der Heatmap als Datei
plt.show()

```
